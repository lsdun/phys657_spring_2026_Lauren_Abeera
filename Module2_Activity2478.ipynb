### Activity 2: Poisson distribution - mean and variance

#### Consider a Poisson process with a rate parameter $\lambda$. The probability of observing k events in a fixed interval is given by the Poisson distribution. 

##### 1. Write down the Poisson probability mass function $p(k \mid \lambda)$.

The probability mass function for observing \( k \) events in a fixed interval is

$$
p(k \mid \lambda) = \frac{\lambda^k e^{-\lambda}}{k!},
$$

where \( k = 0, 1, 2,... \).


##### 2. Compute the expectation value $<k>$.

The expectation value of \( k \) for a Poisson distribution is defined as

$$
\langle k \rangle = \sum_{k=0}^{\infty} k \, p(k \mid \lambda).
$$

Plugging in the probability mass function $ p(k \mid \lambda) $:

$$
\langle k \rangle
= \sum_{k=0}^{\infty} k \frac{\lambda^k e^{-\lambda}}{k!}.
$$

The $ k = 0 $ term vanishes:

$$
\langle k \rangle
= \sum_{k=1}^{\infty} k \frac{\lambda^k e^{-\lambda}}{k!}.
$$

Using the identity

$$
\frac{k}{k!} = \frac{1}{(k-1)!},
$$

we obtain:

$$
\langle k \rangle
= e^{-\lambda} \sum_{k=1}^{\infty} \frac{\lambda^k}{(k-1)!}.
$$

Factoring out a power of $ \lambda $:

$$
\langle k \rangle
= \lambda e^{-\lambda} \sum_{k=1}^{\infty} \frac{\lambda^{k-1}}{(k-1)!}.
$$

Letting $ j = k - 1 $, this becomes:

$$
\langle k \rangle
= \lambda e^{-\lambda} \sum_{j=0}^{\infty} \frac{\lambda^j}{j!}.
$$

Using the Taylor series expansion of the exponential function:

$$
\sum_{j=0}^{\infty} \frac{\lambda^j}{j!} = e^{\lambda},
$$

we are left with:

$$
\langle k \rangle = \lambda e^{-\lambda} e^{\lambda} = \lambda.
$$

Therefore,

$$
\boxed{\langle k \rangle = \lambda.}
$$


##### 3. Compute the variance $\mathrm{Var}(k)$.

The variance of k is given by **Bishop Eq. 1.51**:

$$
\mathrm{Var}(k) = \langle k^2 \rangle - \langle k \rangle^2.
$$

For a Poisson distribution with parameter $ \lambda $, we just calculated the expectation value to be:

$$
\langle k \rangle = \lambda.
$$

Now computing $ \langle k^2 \rangle $:

$$
\langle k^2 \rangle
= \sum_{k=0}^{\infty} k^2 \, p(k \mid \lambda)
= \sum_{k=0}^{\infty} k^2 \frac{\lambda^k e^{-\lambda}}{k!}.
$$

Rewriting  $ \langle k^2 \rangle $ as $k^2 = k(k-1) + k$:

$$
\langle k^2 \rangle
= \sum_{k=0}^{\infty} k(k-1)\frac{\lambda^k e^{-\lambda}}{k!}
+ \sum_{k=0}^{\infty} k\frac{\lambda^k e^{-\lambda}}{k!}.
$$

The second sum is simply

$$
\langle k \rangle = \lambda.
$$

For the first sum, note that $ \frac{k(k-1)}{k!} = \frac{1}{(k-2)!}$ and the terms with \( k=0,1 \) vanish. Thus:

$$
\sum_{k=0}^{\infty} k(k-1)\frac{\lambda^k e^{-\lambda}}{k!}
= e^{-\lambda} \sum_{k=2}^{\infty} \frac{\lambda^k}{(k-2)!}.
$$

Factor out $ \lambda^2 $:

$$
= \lambda^2 e^{-\lambda} \sum_{k=2}^{\infty} \frac{\lambda^{k-2}}{(k-2)!}.
$$

Letting $ j = k - 2 $, this becomes:

$$
= \lambda^2 e^{-\lambda} \sum_{j=0}^{\infty} \frac{\lambda^j}{j!}
= \lambda^2.
$$

Therefore,

$$
\langle k^2 \rangle = \lambda^2 + \lambda.
$$

Substituting into the variance:

$$
\mathrm{Var}(k)
= \langle k^2 \rangle - \langle k \rangle^2
= (\lambda^2 + \lambda) - \lambda^2
= \lambda.
$$

Therefore,

$$
\boxed{\mathrm{Var}(k) = \langle k \rangle.}
$$


##### 4. Show explicitly that $\mathrm{Var}(k)$ = $\langle k \rangle$.

Start with the Poisson pmf:

$$
p(k\mid \lambda)=\frac{\lambda^k e^{-\lambda}}{k!}, \qquad k=0,1,2,\dots
$$

The expectation value is

$$
\langle k\rangle=\sum_{k=0}^{\infty} k\,p(k\mid\lambda)
= e^{-\lambda}\sum_{k=1}^{\infty} k\frac{\lambda^k}{k!}
= e^{-\lambda}\sum_{k=1}^{\infty}\frac{\lambda^k}{(k-1)!}.
$$

Factor out one power of $\lambda$ and shift the index \(j=k-1\):

$$
\langle k\rangle
= \lambda e^{-\lambda}\sum_{k=1}^{\infty}\frac{\lambda^{k-1}}{(k-1)!}
= \lambda e^{-\lambda}\sum_{j=0}^{\infty}\frac{\lambda^j}{j!}
= \lambda.
$$

Now compute $\langle k(k-1)\rangle$:

$$
\langle k(k-1)\rangle
= \sum_{k=0}^{\infty} k(k-1)p(k\mid\lambda)
= e^{-\lambda}\sum_{k=2}^{\infty} k(k-1)\frac{\lambda^k}{k!}.
$$

Use the identity

$$
\frac{k(k-1)}{k!}=\frac{1}{(k-2)!},
$$

so

$$
\langle k(k-1)\rangle
= e^{-\lambda}\sum_{k=2}^{\infty}\frac{\lambda^k}{(k-2)!}
= \lambda^2 e^{-\lambda}\sum_{k=2}^{\infty}\frac{\lambda^{k-2}}{(k-2)!}.
$$

Let $j=k-2$:

$$
\langle k(k-1)\rangle
= \lambda^2 e^{-\lambda}\sum_{j=0}^{\infty}\frac{\lambda^j}{j!}
= \lambda^2.
$$

Next, use the identity $k^2 = k(k-1)+k$ to write

$$
\langle k^2\rangle = \langle k(k-1)\rangle + \langle k\rangle
= \lambda^2 + \langle k\rangle.
$$

Now compute the variance:

$$
\mathrm{Var}(k)=\langle k^2\rangle-\langle k\rangle^2
$$

$$
= (\lambda^2+\langle k\rangle)-\langle k\rangle^2.
$$

From above, $\langle k\rangle=\lambda$, so $\lambda^2=\langle k\rangle^2$. Substituting gives

$$
\mathrm{Var}(k)=\langle k\rangle^2+\langle k\rangle-\langle k\rangle^2=\langle k\rangle.
$$

Therefore,

$$
\boxed{\mathrm{Var}(k)=\langle k\rangle.}
$$


##### 5. Briefly discuss the physical meaning of this result for counting experiments.

In a Poisson distribution, the statistical fluctuation is directly determined by the mean number of counts for counting experiments. Because the standard deviation is equal to the square root of the mean ($\sigma$ = $\sqrt \lambda$), higher counts result in higher absolute vairation but lower relative error. Larger total counts will yield more precise measurements. This property specifically applies to events that occur at a constant average rate, such as radioactive decay or particle detections. 

### Activity 4: Maximum Likelihood Estimation (by hand) for a Gaussian

#### Assume a dataset $ \{x_n\}_{n=1}^{N} $ is drawn independently from a Gaussian distribution with unknown mean $ \mu $ and variance $ \sigma^2 $.

##### 1. Write down the likelihood function $ p(\{x_n\} \mid \mu, \sigma^2) $.


Because the observations are independently and identically distributed, the joint probability of the entire data set is the product of the individual Gaussian probabilities
**(Bishop Eq. 1.53)**:

$$
p(\mathbf{x} \mid \mu, \sigma^2)
= \prod_{n=1}^{N} \mathcal{N}(x_n \mid \mu, \sigma^2).
$$

Now the explicit Gaussian density can be substituted for
$ \mathcal{N}(x_n \mid \mu, \sigma^2) $.
For the univariate Gaussian, the distribution is given by **Bishop Eq. 2.218**:

$$
p(x \mid \mu, \sigma^2)
= \frac{1}{(2\pi\sigma^2)^{1/2}}
\exp\!\left\{-\frac{1}{2\sigma^2}(x-\mu)^2\right\}.
$$

The resulting likelihood function is therefore

$$
p(\mathbf{x} \mid \mu, \sigma^2)
= \prod_{n=1}^{N}
\frac{1}{(2\pi\sigma^2)^{1/2}}
\exp\!\left\{-\frac{(x_n-\mu)^2}{2\sigma^2}\right\}.
$$



##### 2. Write down the log-likelihood.

Starting from the likelihood

$$
p(\mathbf{x} \mid \mu, \sigma^2)
= \prod_{n=1}^{N}
\frac{1}{(2\pi\sigma^2)^{1/2}}
\exp\!\left\{-\frac{(x_n-\mu)^2}{2\sigma^2}\right\},
$$

the log-likelihood is

$$
\ln p(\mathbf{x} \mid \mu, \sigma^2)
= \ln \prod_{n=1}^{N}
\left[
\frac{1}{(2\pi\sigma^2)^{1/2}}
\exp\!\left\{-\frac{(x_n-\mu)^2}{2\sigma^2}\right\}
\right].
$$

Using $\ln \prod_n x_n = \sum_n \ln x_n$, we obtain

$$
\ln p(\mathbf{x} \mid \mu, \sigma^2)
= \sum_{n=1}^{N}
\ln \left[
\frac{1}{(2\pi\sigma^2)^{1/2}}
\exp\!\left\{-\frac{(x_n-\mu)^2}{2\sigma^2}\right\}
\right].
$$

Using log properties: 

$$
\ln p(\mathbf{x} \mid \mu, \sigma^2)
= \sum_{n=1}^{N}
\left[
\ln\!\left(\frac{1}{(2\pi\sigma^2)^{1/2}}\right)
+ \ln\!\left(\exp\!\left\{-\frac{(x_n-\mu)^2}{2\sigma^2}\right\}\right)
\right].
$$

$$
\ln p(\mathbf{x} \mid \mu, \sigma^2)
= \sum_{n=1}^{N}
\left[
-\frac{1}{2}\ln(2\pi\sigma^2)
-\frac{(x_n-\mu)^2}{2\sigma^2}
\right].
$$

There is no index in the first term so the log-likelihood simplifies to:

$$
\ln p(\mathbf{x} \mid \mu, \sigma^2)
= -\frac{N}{2}\ln(2\pi\sigma^2)
-\frac{1}{2\sigma^2}\sum_{n=1}^{N}(x_n-\mu)^2.
$$


##### 3. Derive the maximum likelihood estimator for $\mu$.

Take the derivative of log-likelihood with respect to $\mu$. Only the second term depends on $\mu$:

$$
\frac{\partial}{\partial \mu}\ln p(\mathbf{x} \mid \mu, \sigma^2)
$$

$$
= -\frac{1}{2\sigma^2}\frac{\partial}{\partial \mu}\sum_{n=1}^{N}(x_n-\mu)^2.
$$

Differentiate inside the sum:

$$
\frac{\partial}{\partial \mu}(x_n-\mu)^2 = -2(x_n-\mu).
$$

Therefore:

$$
\frac{\partial}{\partial \mu}\ln p(\mathbf{x} \mid \mu, \sigma^2)
= -\frac{1}{2\sigma^2}\sum_{n=1}^{N}\left[-2(x_n-\mu)\right]
= \frac{1}{\sigma^2}\sum_{n=1}^{N}(x_n-\mu).
$$

Set derivative equal to zero:

$$
\frac{1}{\sigma^2}\sum_{n=1}^{N}(x_n-\mu)=0
$$
$$
\sum_{n=1}^{N}x_n - N\mu = 0.
$$

Solve for $ \mu $:

$$
\hat{\mu} = \frac{1}{N}\sum_{n=1}^{N}x_n.
$$

Thus, the maximum likelihood estimator for the mean is

$$
\boxed{\hat{\mu} = \frac{1}{N}\sum_{n=1}^{N}x_n.}
$$


##### 4. Derive the maximum likelihood estimator for $\sigma ^2$.

Starting from the log-likelihood for a Gaussian,

$$
\ln p(\mathbf{x} \mid \mu, \sigma^2)
= -\frac{N}{2}\ln(2\pi\sigma^2)
-\frac{1}{2\sigma^2}\sum_{n=1}^{N}(x_n-\mu)^2,
$$

Differentiate the log-likelihood with respect to $\sigma^2$:

$$
\frac{\partial}{\partial \sigma^2}\ln p(\mathbf{x} \mid \mu, \sigma^2)
= -\frac{N}{2}\frac{1}{\sigma^2}
+ \frac{1}{2(\sigma^2)^2}\sum_{n=1}^{N}(x_n-\mu)^2.
$$

Set the derivative equal to zero:

$$
-\frac{N}{2}\frac{1}{\sigma^2}
+ \frac{1}{2(\sigma^2)^2}\sum_{n=1}^{N}(x_n-\mu)^2 = 0.
$$

$$
- N\sigma^2 + \sum_{n=1}^{N}(x_n-\mu)^2 = 0.
$$

$$
\hat{\sigma}^2
= \frac{1}{N}\sum_{n=1}^{N}(x_n-\mu)^2.
$$

Replacing $ \mu $ with its maximum likelihood estimate:
$$
\boxed{
\hat{\sigma}^2
= \frac{1}{N}\sum_{n=1}^{N}\bigl(x_n-\hat{\mu}\bigr)^2
}.
$$


##### 5. Compare your expression for $\sigma ^2$ with the unbiased sample variance estimator. Briefly explain the difference, and discuss when this difference is relevant. 

The maximum likelihood estimator for the variance is

$$
\hat{\sigma}^2 = \frac{1}{N}\sum_{n=1}^{N}(x_n-\hat{\mu})^2,
$$

where $ \hat{\mu} = \frac{1}{N}\sum_{n=1}^{N}x_n $.
The unbiased sample variance estimator is given by **Bishop Eq. 1.59**:

$$
\sigma^2 = \frac{1}{N-1}\sum_{n=1}^{N}(x_n-\mu)^2.
$$

The difference between these two estimators is the factor of $N-1$ in the denominator of the unbiased estimate.
The maximum likelihood estimator divides by $ N $, which results in a slight
downward bias because the same data are used to estimate both the mean and the
variance.
The unbiased estimator divides by $ N-1 $, correcting for the loss of one
degree of freedom associated with estimating the mean. This also removes the bias of maximum liklihood.


### Activity 7: Bayesian inference for Gaussian parameters

##### A. $\sigma ^2$ is known, $\mu$ is inferred. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import matplotlib.colors as mcolors

# Generating Gaussian data set

# Parameters
mu = 0.8 # from figure 2.12
variance = 0.1 # from figure 2.12
sigma = np.sqrt(variance)
N = 1000

# Generate data
x = np.random.normal(mu, sigma, N)

# prior choice
mu_0 = 0.0 
sigma_0 = 0.1 # setting variance to the true value 
```

From **Bishop Eq. 2.143**, the sample mean, or maximum likelihood estimate $\mu_{ML}$ is given by:

$$
\bar{x}_N = \frac{1}{N}\sum_{n=1}^{N}(x_n).
$$

The posterior variance is given by **Bishop Eq. 2.142**:

$$
\frac{1}{\sigma^2_N} = \frac{1}{\sigma^2_0}+\frac{N}{\sigma^2}
$$

The posterior mean of the Gaussian mean parameter is given by

$$
\mu_N
= \sigma_N^2
\left(
\frac{\mu_0}{\sigma_0^2}
+
\frac{N \bar{x}_N}{\sigma^2}
\right).
$$

These equations are defined in the bayesian_inference_mu function.

```python
# Compute the posterior distribution of the mean

def bayesian_inference_mu(x_N, variance, mu_0, sigma_0):

    N = len(x_N) # Number of obsered data points 
    x_bar = np.mean(x_N) # mu_ML

    # posterior variance
    sigma_N = 1.0 / (1.0 / sigma_0 + N / variance)

    # posterior mean
    mu_N = sigma_N * (mu_0 / sigma_0 + N * x_bar / variance)

    return mu_N, sigma_N, x_bar

# Compare for different N from Figure 2.12

N_values = [1, 2, 10]

results = []
for N in N_values:
    x_N = x[:N]
    mu_N, sigma_N, mu_ML = bayesian_inference_mu(x_N, variance, mu_0, sigma_0)
    results.append((N, mu_N, sigma_N, mu_ML))

# Plot prior and posteriors

mu_grid = np.linspace(-1, 2, 600)

plt.figure(figsize=(9, 5))

# Prior (N = 0)
plt.plot(
    mu_grid,
    norm.pdf(mu_grid, mu_0, np.sqrt(sigma_0)),
    color = "orangered",
    label="Prior (N = 0)"
)

colors = {
    1: "hotpink",
    2: "royalblue",
    10: "forestgreen"
}

for N, mu_N, sigma_N, mu_ML in results:
    plt.plot(
        mu_grid,
        norm.pdf(mu_grid, mu_N, np.sqrt(sigma_N)),
        color=colors[N],
        label=f"Posterior (N = {N})"
    )

# True mean
plt.axvline(mu, linestyle="--", color="k", label="True ")

plt.xlabel("")
plt.ylabel("Density")
plt.title("Bayesian inference of  with known variance")
plt.legend()
plt.show()
```

##### B. $\mu$ is known, $\sigma ^2$ is inferred. 

Following **Bishop Sec. 2.3.6**, it is convenient to work with the precision $\lambda = \frac{1}{\sigma^2}$.

For a Gaussian likelihood with known mean, the likelihood as a function of $ \lambda $ takes
the form

$$
p(\mathbf{X} \mid \lambda)
\propto
\lambda^{N/2}
\exp\!\left(
-\frac{\lambda}{2}\sum_{n=1}^{N}(x_n-\mu)^2
\right).
$$

The conjugate prior for this likelihood is a Gamma distribution,

$$
p(\lambda) = \mathrm{Gam}(\lambda \mid a_0, b_0)
= \frac{1}{\Gamma(a_0)} b_0^{a_0}\lambda^{a_0-1} e^{-b_0\lambda},
$$
where $ a_0 $ and $ b_0 $ are the prior shape and rate parameters.

Multiplying the likelihood by the prior yields a posterior distribution that is also Gamma,
$$
p(\lambda \mid \mathbf{X}) = \mathrm{Gam}(\lambda \mid a_N, b_N),
$$
with updated parameters
$$
a_N = a_0 + \frac{N}{2},
\qquad
b_N = b_0 + \frac{1}{2}\sum_{n=1}^{N}(x_n-\mu)^2.
$$

Thus, Bayesian inference for the variance reduces to updating the parameters of a Gamma
distribution as data are observed. 

Because $ \sigma^2 = 1/\lambda $, the posterior mean of the variance is
$$
\mathbb{E}[\sigma^2\mid \mathbf{X}] = \mathbb{E}\!\left[\frac{1}{\lambda}\right]
= \frac{b_N}{a_N-1},
\qquad a_N>1.
$$

The maximum likelihood estimator of the variance (when $ \mu $ is known) is
$$
\hat{\sigma}^2_{\mathrm{ML}} = \frac{1}{N}\sum_{n=1}^{N}(x_n-\mu)^2
= \frac{S}{N}.
$$

where we define $ S \equiv \sum_{n=1}^{N}(x_n-\mu)^2. $

The corresponding ML estimate of precision is
$$
\hat{\lambda}_{\mathrm{ML}} = \frac{1}{\hat{\sigma}^2_{\mathrm{ML}}}.
$$

These equations are defined in the bayesian_lambda function.


```python
from scipy.stats import gamma

a0 = 2.0 # shape parameter for gamma
b0 = 2.0 # rate parameter for gamma

def bayesian_lambda(x_N, mu_known, a0, b0):

    N = len(x_N)
    S = np.sum((x_N - mu_known)**2)

    aN = a0 + 0.5 * N
    bN = b0 + 0.5 * S

    # Posterior mean of lambda 
    lambda_mean = aN / bN

    # Posterior mean of sigma^2
    sigma2_mean = bN / (aN - 1.0) if aN > 1 else np.nan

    # ML estimate of sigma^2 
    sigma2_ML = S / N

    return aN, bN, lambda_mean, sigma2_mean, sigma2_ML

lambda_results = []
for N in N_values:
    x_N = x[:N]
    aN, bN, lam_mean, sig2_mean, sig2_ML = bayesian_lambda(x_N, mu, a0, b0)
    lambda_results.append((N, aN, bN, lam_mean, sig2_mean, sig2_ML))

# Plot prior and posteriors for lambda

lambda_grid = np.linspace(0.01, 12, 1000)

plt.figure(figsize=(9, 5))

# Prior (N=0)
plt.plot(
    lambda_grid,
    gamma.pdf(lambda_grid, a=a0, scale=1.0/b0),
    color = "orangered",
    label="Prior (N = 0)"
)

# Posteriors
for (N, aN, bN, lam_mean, sig2_mean, sig2_ML) in lambda_results:
    plt.plot(
        lambda_grid,
        gamma.pdf(lambda_grid, a=aN, scale=1.0/bN),
        color=colors[N],
        label=f"Posterior (N = {N})"
    )

lambda_true = 1.0 / variance
plt.axvline(lambda_true, linestyle="--", color="k", label="True ")

plt.xlabel(r"$\lambda$")
plt.ylabel("density")
plt.title(r"Bayesian inference of precision $\lambda = 1/\sigma^2$ with known $\mu$")
plt.legend()
plt.show()
```

##### C. Explain the idea of a conjugate prior and how you can develop algorithms to implement Bayesian inference with an iterative process.


- In Bayesian inference, unknown parameters are treated as random variables and assigned prior distributions that represent beliefs before observing data. A conjugate prior is chosen so that the posterior distribution has the same functional form as the prior after applying Bayes theorem, meaning the posterior and prior distributions belong to the same distribution. This allows Bayesian updates to be performed analytically by updating a small number of parameters. Otherwise, evaluating the posterior, which often requires integration over products of likelihoods and priors, can be computationally expensive or may not have a closed-form solution. 

- Conjugacy allows for an iterative inference algorithm:
  - Start with an initial prior.
    - For a Gaussian likelihood with known variance and unknown mean, a Gaussian prior on the mean is conjugate (results in a Gaussian posterior whose mean and variance can be updated analytically).
    - For a Gaussian likelihood with known mean and unknown variance, it is convienent to work with the precision $\lambda$ = $1/\sigma ^2$, for which a Gamma prior is conjugate.
  - Combine prior belief with the information from the data to produce a posterior.
  - Use the posterior as the new prior when additional data are observed.

- This approach allows parameter estimates to be updated efficiently using only a few summary statistics (e.g. sample mean, number of points, sum of squared deviations). As more observations are incorporated, the posterior distribution becomes more concentrated and the influence of the initial prior diminishes. Bayesian estimates converging to the maximum likelihood solution in the large-sample limit.

#### 4. Discuss how the influence of the prior changed as $N$ increases.

As the number of observations increases, the contribution from the likelihood grows proportionally
to $N$. The contribution from the prior remains fixed because the prior represents beliefs before seeing the data. Consequently, as $ N \rightarrow \infty $, the likelihood function overwhelms the prior, making the initial choice of prior width basically irrelevant. Since the posterior distribution is proportional to the product of the likelihood and the prior, the posterior mean approaches the maximum likelihood estimate, regardless of the initial prior width.

For small numbers of observations, the posterior distribution's shape, mean, and variance are strongly influenced by the prior. A narrow prior, which corresponds to a small prior variance, causes the posterior mean to shrink towards the prior mean and results in a smaller posterior variance, while a broad prior allows the data to play a larger role in the result even for small $N$. This is how Bayesian inference provides informed estimates for both large and small sample sizes.
